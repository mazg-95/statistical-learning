### Modelo Regresion Lineal Multiple

```python
def compute_logits(X, params):
    ## add a column of one's to add the bias (b) in the same vectorize operation
    with tf.name_scope("Logits"):
        padding = tf.constant([[0,0],[1, 0]], name="Padding")
        X_with_bias = tf.pad(X, padding, "CONSTANT", constant_values=1, name="PreparedData")
        logits = tf.linalg.matmul(X_with_bias, params, name="Logits")
    return logits

def build_predict(params):
    return lambda x: tf.nn.softmax(compute_logits(x, params), name="Model")

def compute_error(y, logits):
    with tf.name_scope("Cost"):
        cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels = y, logits = logits, name="CrossEntropy")
        error = tf.reduce_mean(cross_entropy, name="Error")
    return error

def compute_accuracy(y, y_hat):
    with tf.name_scope("accuracy"):
        labels_predicted = tf.argmax(y_hat, axis=1)
        labels = tf.argmax(y, axis = 1)
        matched = tf.dtypes.cast(tf.equal(labels_predicted, labels), tf.int32)
        accuracy = tf.divide(tf.reduce_sum(matched), tf.shape(y)[0], name="Computeaccuracy")
    return accuracy
    

def build_linear_model_graph(tensor_x, tensor_y, params, lr=0.001):
    logits = compute_logits(tensor_x, params)
    predict = build_predict(params)
    error = compute_error(tensor_y, logits)
    error_summary = tf.summary.scalar(name="ErrorSummary", tensor=error)
    accuracy = compute_accuracy(tensor_y, predict(tensor_x))
    accuracy_summary = tf.summary.scalar(name="accuracySummary", tensor=accuracy)
    with tf.name_scope("GradieDescent"):
        gradient = tf.gradients(error, [params], name="Gradient")
        update_params = tf.assign(params, params - lr * gradient[0], name="UpdateParams")
    return logits, update_params, params, error, error_summary, accuracy, accuracy_summary, predict

def log(theta, epoch, error, accuracy):
    print(f"MB Gradient Descent_epoch_{epoch}_error_{error}_accuracy_{accuracy}")

def fit(X, Y, batch_size=32, epochs=100, lr=0.001, print_n=10, logs_path='./logs/graph_'):
    path = f"{logs_path}_lr_{lr}_batch_size_{batch_size}_{datetime.today().strftime('%Y%m%d-%H%M%S')}"
    tf.reset_default_graph()
    # Parameters Matrix [features + 1, labels] (+1 to add the weight for the bias X0)
    params = tf.get_variable("Params", dtype=tf.float32, shape=[X.shape[1] + 1, Y.shape[1]],initializer=tf.zeros_initializer())
    # shape: (?, features)
    tensor_x = tf.placeholder(tf.float32, [None,X.shape[1]], name="X")
    # shape: (?, labels)
    tensor_y = tf.placeholder(tf.float32, [None,Y.shape[1]], name="Y")
    _, update_params, params, error, error_summary, accuracy, accuracy_summary, predict = build_linear_model_graph(tensor_x, tensor_y, params, lr)
    n = X.shape[0] // batch_size
    final_tetha = None
    with tf.train.MonitoredSession() as session:
        with tf.summary.FileWriter(path, session.graph) as train_writer:
            for epoch in range(epochs): 
                for i in range(n):
                    start, end = i * batch_size, (i+1) * batch_size
                    x_mbatch, y_mbatch = X[start:end,], Y[start:end,]
                    feed_dict = { tensor_x: x_mbatch, tensor_y: y_mbatch}
                    _, theta = session.run([update_params, params], feed_dict=feed_dict)
                if (epoch + 1) % print_n == 0:
                    feed_dict = { tensor_x: X, tensor_y: Y }
                    val_error, val_accuracy, e_summary,acc_summary, theta =  session.run([error, accuracy, error_summary, accuracy_summary, params], feed_dict = feed_dict)
                    train_writer.add_summary(e_summary, epoch + 1)
                    train_writer.add_summary(acc_summary, epoch + 1)
                    log(theta, epoch + 1, val_error, val_accuracy)
            final_theta = session.run(params, feed_dict=feed_dict)
    return final_theta, predict
```

```python
def run(x_train, y_train, x_test, y_test, batch_size=32, epochs=100, lr=0.001, print_n=10, logs_path='./logs/graph_'):
    theta, predict = fit(x_train, y_train, epochs=epochs, lr=lr, print_n=print_n, logs_path=logs_path)
    # shape: (?, features)
    tensor_x_test = tf.placeholder(tf.float32, [None,x_test.shape[1]], name="X_test")
    # shape: (?, labels)
    tensor_y_test = tf.placeholder(tf.float32, [None,y_test.shape[1]], name="Y_test")

    get_predictions = build_predict(theta)(tensor_x_test)
    get_accuracy = compute_accuracy(tensor_y_test, get_predictions)
    with tf.Session() as session:
        with tf.summary.FileWriter(logs_path+"prediction", session.graph) as train_writer:
            y_estimates, accuracy = session.run([get_predictions, get_accuracy], feed_dict = { tensor_x_test: x_test, tensor_y_test: y_test})
    return y_estimates, accuracy
    
```

```python
y_hat, accuracy = run(train_images_processed, train_labels_processed, test_images_processed, test_labels_processed, epochs=1000, lr=0.0001, logs_path='./logs_mnist/graph_')
print(accuracy)
```

```python
y_hat, accuracy = run(train_images_scaled, train_labels_processed, test_images_scaled, test_labels_processed, epochs=1000, lr=0.0001, logs_path='./logs_mnist/graph_')
print(accuracy)
```

```python
y_hat, accuracy = run(train_images_normalized, train_labels_processed, test_images_normalized, test_labels_processed, epochs=1000, lr=0.0001, logs_path='./logs_mnist/graph_')
print(accuracy)
```
